I used the pdict command to pull out the data definition of a conetic table and imported that information to a postgresql table so that I could produce the column lists needed to export all the data.

```python
import psycopg2
from psycopg2 import sql

# Database connection parameters
db_params = {
    "host": "localhost",
    "database": "landy",
    "user": "kristoferpearson",
    "password": ""
}

# Establish a connection to the database
conn = psycopg2.connect(**db_params)

# Create a cursor object
cur = conn.cursor()

# Execute a query
query = sql.SQL("SELECT * FROM cpa_gastate_pdict ORDER BY id;")
cur.execute(query)

# Fetch all rows
rows = cur.fetchall()

# Get column names
column_names = [desc[0] for desc in cur.description]

field_list = ""
header = ""
print_list = ""
count = 0

# Iterate over the rows
for row in rows:
    # Create a dictionary with column names as keys
    row_dict = dict(zip(column_names, row))
    
    count += 1
    field_list = field_list + f'{row_dict['col_name']}, '

    if not count % 6:
        field_list = field_list + '\n'

    if row_dict['quantity'] == 1:
        header = header + f'PRINT "{row_dict['col_name']}^";\n'
        print_list = print_list + f'PRINT {row_dict['col_name']}, "^";\n'
    else:
        for i in range(row_dict['quantity']):
            header = header + f'PRINT "{row_dict['col_name']}[{i+1}]^";\n'
            print_list = print_list + f'PRINT {row_dict['col_name']}[{i+1}], "^";\n'

# Close the cursor and connection
cur.close()
conn.close()

# Print out lists for Grace Script
print(f"field_list = {field_list}")
print(f"header = {header.rstrip()}")
print(f"print_list = {print_list.rstrip()}")
```

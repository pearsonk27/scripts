```python
import re

regex = r'^(?P<columnName>.*):(?P<dataType>.*):(?P<size>.*):(?P<quantifier>.*)$'

data_types = []

with open('/Users/kristoferpearson/Dev/scripts/python/Landy/pull-conetic-def/lawyer_pdict.txt', "r", encoding = "utf-8") as lawyer_pdict_file:
    for line in lawyer_pdict_file.read().splitlines():
        match = re.match(regex, line)
        if match:
            data_type = {}
            data_type['column_name'] = match.group('columnName')
            data_type['data_type'] = match.group('dataType')
            data_type['size'] = int(match.group('size'))
            data_type['quantifier'] = int(match.group('quantifier'))
            data_types.append(data_type)

print("Column List: ")

for data_type in data_types:
    print(f'"{data_type['column_name']}",')

print("Data Type: ")

for data_type in data_types:
    if data_type['data_type'] == 'integer':
        line = f'INT {data_type['column_name']}'
    else:
        line = f'{data_type['data_type'].upper()} {data_type['column_name']}'
    if data_type['quantifier'] > 1:
        line += f'[{data_type['quantifier']}]'
    if data_type['size'] > 1 and data_type['data_type'] not in ['integer', 'date', 'real', 'money']:
        line += f'[{data_type['size']}]'
    line += ';'
    print(line)
```
